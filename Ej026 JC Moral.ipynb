## Ejercicio 1: Secuencia de Fibonacci

Crear una lista con los 30 primeros números de la secuencia de Fibonacci. Esta secuencia comienza con los números 0 y 1, y los números siguientes corresponden a la suma de los dos números anteriores:

0, 1, 1, 2, 3, 5, 8, 13, ...

Antes de crear la secuencia debes crear un diagrama de flujo que explique el algoritmo que vas a programar. 

```python
from PIL import Image
path = "fibojc1.jpg"
display(Image.open(path))
```

```python
# Utilizo iteraciones, es decir:
# x_{n}+ x_{n+1} = x_{n+2}
# luego redefino x_{n}=x_{n+1}
# y x_{n+1}= x_{n+2}

a = 0
b = 1
n = 30 # aqui escojo el numero de elemtos de lista que agregare

fibonacci = [a,b]  # los primeros elemetos de la lista son 0 y 1
# Ahora comienza a iterar: 
for i in range(n):
	c = a+b
	a= b
	b = c
	fibonacci.append(c)

print(fibonacci)
```

## Ejercicio 2: Dominio de correo electrónico

A partir de un correo electrónico cualquiera debes crear un algoritmo que entregue solo el dominio, es decir, quitando el nombre de usuario. 

## Ingrese nombre de usuario

```python
correo= "usuario@dominio.com"
n = correo.find("@")  # obtiene el indice donde se aloja el simbolo @ y lo guarda
dominio = correo[n:] # luego imprimira desde el indice n hasta el final del string
print(dominio)
```
